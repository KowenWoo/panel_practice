```python
import panel as pn
import numpy as np
import hvplot.pandas 
import pandas as pd
pn.extension()

#data
csv_file = ("https://raw.githubusercontent.com/holoviz/panel/main/examples/assets/occupancy.csv")
data = pd.read_csv(csv_file)

#widget
selecty = pn.widgets.Select(options = [
    'Temperature', 'Humidity', 'Light', 'CO2', 'HumidityRatio'
])

#graph
graph = data.hvplot.line(x = 'date', y = selecty)

#allows for graph y axis to be changed through front end
def display_graph(a):
    display(a)
pn.bind(display_graph, selecty)

#organizes data structures into tabs
tabs = pn.Tabs(data)
tabs.extend([
    ("Graph", graph),
    ("Widget", selecty)
])
tabs
```
